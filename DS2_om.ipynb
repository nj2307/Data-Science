# OUTLIERS

```python
# Importing
import sklearn
from sklearn.datasets import load_boston
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore') 
```

```python
# Load the dataset
bos_hou = load_boston()
```

```python
column_name = bos_hou.feature_names
df = pd.DataFrame(bos_hou.data)
df.columns = column_name
df.head()
```

```python
import seaborn as sns

plt.figure(figsize=(10,5))
sns.boxplot(df['DIS'])
```

```python
# Position of the Outlier
print(np.where(df['DIS']>10))
```

```python
df.info()
```

```python
plt.figure(figsize=(12,5))

sns.scatterplot(df.INDUS , df.TAX)
plt.show()
```

```python
# Position of the Outlier
print(np.where((df['INDUS']>20) & (df['TAX']>600)))
```

# Removing Outliers

```python
import numpy as np

Q1 = np.percentile(df['DIS'], 25)
Q3 = np.percentile(df['DIS'], 75)
IQR = Q3 - Q1

print("Old Shape: ", df.shape)

lower_bound = (df['DIS'] >= (Q1 - 1.5 * IQR))
upper_bound = (df['DIS'] <= (Q3 + 1.5 * IQR))

df = df[lower_bound & upper_bound]

print("New Shape: ", df.shape)
```

# Scaling

```python
import seaborn as sns
import pandas as pd
import numpy as np

data = sns.load_dataset('iris')
print('Original Dataset')
data.head()
```

```python
df = data.drop('species', axis=1)
df_norm = (df-df.min())/(df.max()-df.min())
df_norm = pd.concat((df_norm, data.species), 1)

print("Scaled Dataset Using Pandas")
df_norm.head()
```
